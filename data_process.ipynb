```python
#input data type processed_dips[0]['frames'][0]  이런식 
def view_add(a_frame):
    #안에 x,y 좌표 담은거 들어있음
    list_to_return=[]
    for a_view in view_list:
        # ex) a_view=processed_dips[0]['frames'][0]['view1']
        view_pts=[]
        for body in body_list:
            view_pts.append(a_frame[a_view]['pts'][body]["x"])
            view_pts.append(a_frame[a_view]['pts'][body]["y"])
        list_to_return.append(view_pts)
    return list_to_return
    
```

```python
total_data=[]

for a_json in processed_dips:
    one_row=[]
    for a_frame in a_json['frames']:
            one_row.append(view_add(a_frame))
    total_data.append(one_row)
```
